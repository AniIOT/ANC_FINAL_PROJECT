```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
my_data = np.genfromtxt('001.csv', delimiter=',')
```

```python
ACC0 = np.sqrt(my_data[:,0]**2 + my_data[:,1]**2 + my_data[:,2]**2)
```

```python
GYR0 = np.sqrt(my_data[:,3]**2 + my_data[:,4]**2 + my_data[:,5]**2)
```

```python
plt.plot(ACC0,label="Accelerometer")
plt.plot(GYR0,label="Gyroscope")
plt.legend()
plt.show()
```
